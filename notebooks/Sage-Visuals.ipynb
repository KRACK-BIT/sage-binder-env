```python
def fft_reorder_inner(z):
    l = len(z) // 2

    if l == 0:
        yield z[0]
        return

    for l, r in zip(fft_reorder_inner(z[:l]), fft_reorder_inner(z[l:])):
        yield l
        yield r
```

```python
for i in tqdm(range(12)):
    exp = list(fft_reorder_inner(list(range(1 << i))))
    obs = [ZZ(n.digits(2, padto=i)[::-1], base=2) for n in srange(1 << i)]
    assert exp == obs, f"{i=} {exp=}, {obs=}"
```

```python
def zeta_ordered_rec(n, R, must_max=False):
    assert n == 1 << (n.bit_length() - 1), f"n must be a power of 2! {n=}"

    zeta = max(R.zeta(n, all=True)) if must_max else R.zeta(n)
    return fft_reorder_inner([zeta**i for i in range(n)])


from itertools import islice


def nega_zeta_ordered_rec(n, R, must_max=False):
    return islice(zeta_ordered_rec(2 * n, R, must_max), n, None)
```

```python
def zeta_ordered(n, R, must_max=False):
    n_len = n.bit_length() - 1
    assert n == 1 << n_len, f"n must be a power of 2! {n=}"

    zeta = max(R.zeta(n, all=True)) if must_max else R.zeta(n)
    yield from (
        zeta ** (ZZ(ZZ(i).digits(2, padto=n_len)[::-1], base=2)) for i in srange(n)
    )


def nega_zeta_ordered(n, R, must_max=False):
    twist = max(R.zeta(2 * n, all=True)) if must_max else R.zeta(2 * n)
    yield from (zeta * twist for zeta in zeta_ordered(n, R, must_max))
```

```python
cases = 12
F = CyclotomicField(1 << cases)
for i in tqdm(range(cases)):
    exp = list(zeta_ordered_rec(1 << i, F))
    obs = list(zeta_ordered(1 << i, F))
    assert exp == obs, f"{i=} {exp=}, {obs=}"

    exp = list(nega_zeta_ordered_rec(1 << i, F))
    obs = list(nega_zeta_ordered(1 << i, F))
    assert exp == obs, f"{i=} {exp=}, {obs=}"
```

```python
def graph_prod_tree(prod_tree):
    prod_layers = list(reversed(prod_tree.layers))
    return Graph(
        [
            (k, v)
            for k_set, v_set in zip(prod_layers, prod_layers[1:])
            if (vit := iter(v_set)) is not None
            for k, v_subset in zip(k_set, zip(vit, vit))
            for v in v_subset
        ]
    )


# print(graph_prod_tree(Kcrt).tikz(rankdir='up').tikz_picture_code())
# print(graph_prod_tree(Kcrt_fine).tikz(rankdir='up').tikz_picture_code())

# display(
#     graph_prod_tree(Kcrt).plot(layout="tree"),
#     graph_prod_tree(Kcrt_fine).plot(layout="tree"),
# )
```
